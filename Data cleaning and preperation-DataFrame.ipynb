```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.randn(4,3),index=['a','c','d','f'])
columns=['one','two','three']
df=df.reindex(['a','b','c','d','e','f'])
print("Original Dataframe with NaN\n",df)
print("\n")
print("Dropped DataFrame\n",df.dropna())
print("\n")











```

```python
df=pd.DataFrame(np.random.randn(4,3),index=['a','c','d','f'])
columns=['one','two','three']
df=df.reindex(['a','b','c'])
print("Original dataFrame:\n",df)
print("\n")
print("NaN replaced with '0':\n",df)
print(df.fillna(0))

```

```python
df=pd.DataFrame({'one':[10,20,80,40,50],'two':[60,70,80,0,10]})
print("Original DataFrame:\n",df)
print("\n DataFrame with replaced values:\n")
print(df.replace({10:5,80:3}))
```

```python
df=pd.DataFrame(np.random.randn(4,3),index=['a','c','d','f'],columns=['one','two','three'])
df=df.reindex(['a','b','c','d','e','f'])
print("/nNaN-with True fill:\n")
print(df['one'].isnull())
```

```python
df=pd.DataFrame(np.random.randn(4,3),index=['a','c','d','f'])
columns=['one','two','three']
print("Original DataFrame:\n",df)
df=df.reindex(['a','b','c','d','e','f'])
print("\nNaN-filled with background values:\n")
print(df.fillna('obj.ffill'))
```
