```
import pandas as pd
df=pd.DataFrame
df=pd.DataFrame()
print(df)
```

```
emp=pd.Series(['parker','john','smith','william'])
id=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id}
result=pd.DataFrame(frame)
print("\nSeries to Data frame\n")
print(result)
```

```
print("\nExtracting one column:\n")
print(result['Emp'])
```

```
print("\n Adding one column:\n")
result ['Age']=pd.Series([35,24,40.38])
print(result)
```

```
print("\n Deleting one column:\n")
del result['Age']
print(result)
```

```
print("\n Extracting the third row:\n")
print(result.loc[2])
```

```
print("slice rows:\n",result[1:3])
```

```
d2=pd.DataFrame([['Dale',123],['Mark',143]],columns=['Emp','ID'])
result=pd.concat([result,d2],ignore_index=True)
print("\n Adding new row values:\n",result)
```

```
print("\n deleting psrticular row:\n",result.drop(1))
```
